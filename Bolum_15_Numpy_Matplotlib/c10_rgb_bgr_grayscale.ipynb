```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = "smile.jpg"

img = cv2.imread(path) #BGR

plt.imshow(img)
plt.show()
```

```python
path = "smile.jpg"

img = cv2.imread(path, 1) #BGR

plt.imshow(img) #rgb
plt.show()
```

```python
path = "smile.jpg"

img = cv2.imread(path, 1) #BGR
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img) #argb
plt.show()
```

```python
path = "smile.jpg"

img = cv2.imread(path, 1) #BGR
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray', interpolation='BICUBIC') #argb
plt.show()
```
