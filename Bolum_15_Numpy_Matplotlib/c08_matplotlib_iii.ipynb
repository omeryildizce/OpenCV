```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(5)
plt.plot(x,[ y**2 for y in x])
plt.plot(x,[ y**3 for y in x])
plt.plot(x,2*x)
plt.plot(x,5.2*x)
plt.grid(True)
print(plt.axis())
plt.show()
```
