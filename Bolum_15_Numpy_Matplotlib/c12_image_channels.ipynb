```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
path = "map.jpeg"

img = plt.imread(path)

plt.imshow(img)
plt.show()
```

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]

output = [img, r, g, b]
titles = ["img","red","green","blue"]

for i in range(4):
    plt.subplot(2,2, i+1)
    plt.axis("off")
    plt.title(titles[i])

    if i == 0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i])
    plt.show()



```
