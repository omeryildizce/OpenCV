```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((500,500), np.uint8) 
```

ravel pikselleri tek bir satıra dönüştürür.

255 kaç değer olduğunu, son parametre değer aralığın 

```python
cv2.imshow("img", img)

plt.hist(img.ravel(), 256, [0,256])  
plt.show()



cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = np.zeros((500,500), np.uint8)  + 50
cv2.rectangle(img,(0,60), (200, 150), (255,255,255), -1)
cv2.rectangle(img,(250,120), (375, 475), (255,255,255), -1)
```

```python
cv2.imshow("img", img)

plt.hist(img.ravel(), 256, [0,256])  
plt.show()



cv2.waitKey(0)
cv2.destroyAllWindows()
```
